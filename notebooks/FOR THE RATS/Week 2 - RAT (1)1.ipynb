## Question 1:
Create a list of numbers and calculate the following using NumPy or Pandas:

## Question 8
Create a line plot of the function y=2x+1. Add a title and axis labels and make the range 0>20  for x and y.
- Hint: Use plt.plot(x, y) 

```python

x = np.arange(0,20,1)
y = x*2
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('y = x*2')
plt.xlabel('x')
plt.ylabel('y')
```
