```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt


# Find project root (folder that contains "data")
def get_project_root():
    p = Path.cwd()
    while not (p / "data").exists() and p != p.parent:
        p = p.parent
    return p

PROJECT_ROOT = get_project_root()
RAW_DATA_DIR = PROJECT_ROOT / "data" / "raw"

def load_data(filename="titanic1.csv"):
    input_path = RAW_DATA_DIR / filename
    print("Reading from:", input_path)  # optional but useful
    return pd.read_csv(input_path)

# Load dataset
goog_data = load_data()
goog_data_np = goog_data.to_numpy()
goog_cp = goog_data_np[:,4]
goog_data
holidays = []
```

```python

goog_data['Fare'] = pd.Series([round(val, 2) for val in goog_data['Fare']],
                              index=goog_data.index)
goog_data['Age'] = pd.Series([round(val, 2) for val in goog_data['Age']],
                              index=goog_data.index)
goog_data['Sex'] = pd.Series([round(val, 2) for val in goog_data['Sex']],
                              index=goog_data.index)

stk_data =goog_data[-10:]
#print(stk_data)
col_head = ["Passengerid", "Age", "Sex"]  
stk_data = goog_data[col_head].tail(10)
#print(col_head)

stk_data_np = stk_data.to_numpy()
plt.figure(linewidth=2,tight_layout={'pad':.5},figsize=(5,3))
axes_8 = plt.gca()
axes_8.get_xaxis().set_visible(False)
axes_8.get_yaxis().set_visible(False)
plt.box(on=None)

# matplotlib.org/3.1.0/tutorials/colors/colormaps.html
ccolors = plt.cm.Blues(np.full(len(col_head), 0.2))

the_table = plt.table(cellText=stk_data_np, loc='center', colLabels=col_head,
                      colColours=ccolors)
the_table.set_fontsize(14)
the_table.scale(3, 2.5)
```
